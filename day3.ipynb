```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [23, 17, 35, 29]

# Create an array for the positions on the x-axis
x_pos = np.arange(len(categories))

# Create the bar graph
plt.bar(x_pos, values, align='center', alpha=0.7)

# Add titles and labels
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph Example')

# Add the category names as x-axis labels
plt.xticks(x_pos, categories)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# frequencies
ages = [2, 5, 70, 40, 30, 45, 50, 45, 43, 40, 44, 60, 7, 13, 57, 18, 90, 77, 32, 21, 40]

# setting the ranges and no. of intervals
age_range = (0, 100)
bins = 10

# plotting a histogram
plt.hist(ages, bins, age_range, color="blue", histtype="bar", edgecolor="black", alpha=0.7)
plt.xlabel('Age')
# frequency label
plt.ylabel('No. of people')
# plot title
plt.title('My Histogram')

# display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# create data for plotting
x_values = [0, 1, 2, 3, 4, 5]
y_values = [0, 1, 4, 9, 16, 25]

# creating the plot
plt.figure()
plt.scatter(x_values, y_values, label="dollar", s=100, color="green")

# adding custom markers ($)
for x, y in zip(x_values, y_values):
    plt.text(x, y, '$', fontsize=12, ha='center', va='center', color='green')

# showing the legend
plt.legend()

# displaying the plot
plt.show()

```
